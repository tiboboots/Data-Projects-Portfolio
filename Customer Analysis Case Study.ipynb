```python
age_ranges = [[20, 29], [30, 39], [40, 49], [50, 59], [60, 69], [70, 80]] 

def assign_age(age):
    for num in age_ranges:
        if num[0] <= age <= num[1]:
            return f'{num[0]} - {num[1]}'
    else:
        return 'NA'
```
