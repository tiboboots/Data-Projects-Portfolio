```python
import pandas as pd
```

```python
movies_df = pd.read_csv('/Users/tibo/Downloads/imdb_movies.csv')
```

```python
movies_df = movies_df.drop('release_date', axis = 1)
```

```python
movies_df = movies_df.drop(['budget_adj', 'revenue_adj'], axis = 1)
```

```python
movies_df = movies_df.drop(['overview', 'tagline', 'homepage', 'imdb_id', 'keywords'], axis = 1)
```

```python
movies_df = movies_df.drop_duplicates()
```

```python
movies_df.dropna()
```

```python
movies_df['profit'] = movies_df['revenue'] - movies_df['budget']
```

```python
movies_df.drop(['cast', 'director', 'production_companies'], axis = 1, inplace = True)
```

```python
movies_df
```

```python
holdmovie = movies_df['genres'].str.split('|', expand = True)
```

```python
holdmovie
```

```python
holdmovie = movies_df['genres'].str.split('|', expand = True).stack()

holdmovie
```

```python
holdmovie.index = holdmovie.index.droplevel(- 1)
```

```python
holdmovie
```

```python
holdmovie.name = 'genres_split'

movies_df = movies_df.join(holdmovie)

movies_df
```

```python
movies_df.drop('genres', axis = 1, inplace = True)
```

```python
movies_df
```

```python
profit_genre = pd.DataFrame(movies_df.groupby(['release_year', 'genres_split'])['profit'].mean().round())
```

```python
profit_genre
```

```python
time_genre = pd.pivot_table(data = profit_genre, values = 'profit', index = 'genres_split', columns = 'release_year')
```

```python
time_genre
```

```python
plt.figure(figsize= [15, 8])

sns.lineplot(data = profit_genre, x = 'release_year', y = 'profit', hue = 'genres_split', style = 'genres_split', markers = True)

plt.ticklabel_format(style = 'plain', axis = 'y')

plt.legend(bbox_to_anchor= (1.2,1))
```

```python
plt.figure(figsize = [20, 10])

sns.heatmap(data = time_genre, cmap = 'coolwarm', linewidths = 2)

plt.tight_layout()
```
